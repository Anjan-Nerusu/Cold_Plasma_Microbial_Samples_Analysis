```python
# Set up the figure size
plt.figure(figsize=(10, 6))

# Custom color palette for different frequencies
b = {
    500: 'blue',
    1000: 'green',
    1500: 'red',
    2000: 'orange',
    2500: 'purple'
}

# Use seaborn to plot a line plot with the custom color palette
sns.lineplot(data=plasma_sd, x='Time_in_minutes', y='Reductions_average', hue='Frequency(Hz)', marker='o', linestyle='-', linewidth=2, palette=b)

# Add labels and title
plt.xlabel('Time(min)')
plt.ylabel('Average log reduction for 3 trails of reduction samples ')
plt.title('Time vs. Average log reductions(R1,R2,R3)')

# Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Custom color palette
c = {
    500: 'blue',
    1000: 'green',
    1500: 'red',
    2000: 'orange',
    2500: 'purple'
}

# Plot the lineplot with custom error bars
plt.figure(figsize=(10, 6))

# Use `yerr` to manually specify error bars from the `Standard_deviation` column
sns.lineplot(data=plasma_sd, x='Time_in_minutes', y='Reductions_average', hue='Frequency(Hz)',
             marker='o', linestyle='-', linewidth=2, palette=c, 
             errorbar=None)  # Disable default error bars

# Now manually add error bars using plt.errorbar()
for a in plasma_sd['Frequency(Hz)'].unique():
    fd = plasma_sd[plasma_sd['Frequency(Hz)'] == a]
    plt.errorbar(fd['Time_in_minutes'], fd['Reductions_average'], 
                 yerr=fd['Standard_deviation'], fmt='o', linestyle='-', 
                 color=c[a], capsize=5)

# Add labels and title
plt.xlabel('Time(min)')
plt.ylabel('Average log reduction for 3 trails of reduction samples')
plt.title('Time vs. Average log reductions (R1, R2, R3) displaying error_bars')

# Add gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
